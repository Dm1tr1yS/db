```python
#Пункт 2
#Функция делит таблицу df на заданное число таблиц(splits)
#Возращает список таблиц
def split_dataframe(df, splits):
    dataframes = []
    index_to_split = len(df)//splits
    start = 0
    end = index_to_split
    for split in range(splits):
        temporary_df = df.iloc[start:end, :]
        dataframes.append(temporary_df)
        start += index_to_split
        end += index_to_split
    return dataframes
```

```python
#Создадим собственную функцию предикат для фильтрации полученных данных
#Пусть функция возвращает true для машин, код выпуска которых попадает в диапазон [min_year,max_year],
#а цена в диапазон [min_price,max_price]  
def year_price_filter(x,params):
    return (params[0]<=x['manufacture_year']<=params[1]) and (params[2]<=x['price_eur']<=params[3])
```

```python
#Теперь создадим функцию для применения наших предикатов
def apply_filter(df,filt):
    filter_df=df
    for i,row in filter_df.iterrows():
        if(not filt(row,[2000,2005,5000,13000])):
            filter_df=filter_df.drop(index=i)
            
    return filter_df
```

```python
#В итоге получили функцию для применения фильтра, которая может принимать на вход любой фильтр.
#Однако параметры фильтра передаются в виде списка параметров
```

```python
#Применим созданный фильтр к таблице с немецкими машинами
filtered_german_car=apply_filter(german_car,year_price_filter)
```
